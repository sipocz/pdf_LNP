```python
import requests
from bs4 import  BeautifulSoup
from os.path import exists

```

```python

def arxiv_pages(code):
    

    header={'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36"}
    url=f"https://arxiv.org/search/?query={code}&searchtype=all&abstracts=hide&order=-announced_date_first&size=50"

    url="https://arxiv.org/list/cs/pastweek?show=1558"
    
    #print(url)
    res=requests.get(url,headers=header)
    
    soup=BeautifulSoup(res.text,"html.parser")
    pdfs=soup.find_all("span",class_="list-identifier")
    pdflist=[]
    for pdf in pdfs:
        st=pdf.a["href"]
        st=st.replace("abs","pdf")
        st="https://arxiv.org"+st+".pdf"
        pdflist.append(st)
    #pages_txt=pages[0].text.strip()
    #print(pages_txt)
 
    return(pdflist)
```

```python
pdflist=arxiv_pages("NLP")
```

```python

def fname_splitter(url):
    import pathlib
    split_up = pathlib.Path(url).parts
    # print(split_up)  # debug
    return(split_up[-1])
 
# fname_splitter('https://arxiv.org/pdf/2202.03914.pdf') # debug function
```
